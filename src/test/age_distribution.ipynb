# Test della distribuzione delle et√†

Genero N persone

```python
from person import Person
from utilities import DataGenerator

generator = DataGenerator()

N = 100000
people = [Person(generator) for _ in range(N)]
```

Creo dei grafici per vedere come sono distribuiti i dati

```python
import matplotlib.pyplot as plt
import numpy as np
from utilities import AgeGroup

ages = [person.age.age_value for person in people]

plt.hist(ages, bins=max(ages), edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of People')
plt.show()

# Dati di esempio (da sostituire con i tuoi)
groups = list(AgeGroup)
intervals = [interval.value[0] for interval in groups]
counts, _ = np.histogram(ages, bins=intervals)
percentages = (counts / N) * 100

# Definizione delle variabili per larghezza e spaziatura delle barre
bar_width = 0.8  # Larghezza delle barre
bar_spacing = 1.5  # Aumenta questo valore per pi√π spazio tra le barre

# Posizioni delle barre, considerando lo spazio aggiuntivo tra di esse
bar_positions = np.arange(len(percentages)) * bar_spacing

# Grafico a barre
plt.bar(bar_positions, percentages, width=bar_width, tick_label=[f'{intervals[i]}-{intervals[i+1]}' for i in range(len(intervals) - 1)])
plt.xlabel('Age Intervals')
plt.ylabel('Percentage')
plt.title('Percentage of People in Each Age Interval')
plt.xticks(rotation=45)  # Rotazione opzionale per etichette leggibili
plt.tight_layout()  # Per migliorare l'estetica complessiva
plt.show()
```
